```python
import matplotlib.pyplot as plt
```

```python
# 도미 데이터
bream_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0,
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0,
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0]
bream_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0,
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0,
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0]
```

```python
plt.scatter(bream_length, bream_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
# 빙어 데이터
smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]
```

```python
plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
#  도미와 빙어 데이터 합치기
length = bream_length + smelt_length
weight =  bream_weight + smelt_weight
```

```python
fish_data = [[l,w] for l,w in zip(length, weight)]
```

```python
#  2차원 리스트
print(fish_data)
```

```python
#  도미는(35마리) 1, 빙어는(14마리) 0으로 표현하기
# 머신러닝에서 찾으려는 대상을 1로 놓고 그외에는 0으로.
fish_target = [1] * 35 +[0] * 14
print(fish_target)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
kn = KNeighborsClassifier()
```

```python
kn.fit(fish_data, fish_target)
```

```python
#  모델 평가 메서도 score() -> 0에서 1사이 값 반환. 1은 모든 데이터를 정확히 맞혔다는 것을 의미.
kn.score(fish_data, fish_target)
```

```python
# k-최근접 이웃 알고리즘
plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.scatter(30,600, marker='^')
plt.scatter(10,20, marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
# 새로운 데이터의 정답을 예측하는 메서드 predict()
kn.predict([[30,600]])
```

```python
kn.predict([[10,20]])
```

```python
#  _fit_X 속성에 fish_data를 가지고 있고
print(kn._fit_X)
```

```python
# _y 속성에 fish_target
print(kn._y)
```

```python
#  KNeighborsClassifier의 기본값인 5 대신 다른 값으로
kn49 = KNeighborsClassifier(n_neighbors=49)
```

```python
# 총 49개데이터 중에 도미가 35개로 다수를 차지하므로 무조건 도미로 예측.
kn49.fit(fish_data, fish_target)
kn49.score(fish_data, fish_target )
```

```python
print(35/49)
```

```python
kn = KNeighborsClassifier()
kn.fit(fish_data, fish_target)

for n in range(5,50):
    # k-최근접 이웃개수 설정
    kn.n_neighbors= n
    #  점수 계산
    score = kn.score(fish_data, fish_target)
    # 100% 정확도에 미치지 못하는 이웃 개수 출력
    if score <1:
        print(n,score)
        break
```

```python
kn17 = KNeighborsClassifier(n_neighbors=17)
kn17.fit(fish_data, fish_target)
kn17.score(fish_data, fish_target )
```
