2. 학습데이터와  테스트 데이터로 분리

```python
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
# X - 2차원이라서 대문자로 표기, y - 1차원이라 소문자로 표기
# train : test = 3:1( test_size = 0.25가 디폴트)
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, random_state=2023)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
#  y_train 값의 분포
np.unique(y_train, return_counts = True)
```

```python
#  y 값의 분포를 균일하게
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, random_state =2023, stratify=iris.target, test_size=0.2)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
#  학습 데이터와 테스트 데이터의 비율(test_size)
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, random_state=2023, stratify=iris.target, test_size =0.2)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
np.unique(y_train, return_counts =True)
```

3. 학습(훈련, train)

```python
#  결정트리모델
from sklearn.tree import DecisionTreeClassifier
```

```python
#  모델 생성,- 객체 생성
dtc = DecisionTreeClassifier()
```

```python
# 하이퍼 파라미터
dtc.get_params()
```

```python
#  학습(훈련)실행
dtc.fit(X_train, y_train)
```

4. 예측(Prediction)

```python
#  X_test 값을 주고 예측값을 반환 받음
pred = dtc.predict(X_test)
```

```python
rf = pd.DataFrame({'y 실제값': y_test, 'y 예측값': pred})
rf.head()
```

```python
X_test[:5]
```

5.평가(Evaluation)

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred)
```

```python
#  4,5 과정을 한번에
dtc.score(X_test, y_test)
```
