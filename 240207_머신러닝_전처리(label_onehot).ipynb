데이터 전처리
1. 레이블 인코딩

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
items = ['TV','냉장고','전자렌지','컴퓨터','선풍기','선풍기','믹서','믹서']
```

```python
# 학습
le.fit(items)
```

```python
# 변환
labels = le.transform(items)
labels
```

```python
# 단축형태(학습, 변환을 한번에)
labels = le.fit_transform(items)
labels
```

```python
#  고수들은 아래같이
labels = LabelEncoder().fit_transform(items)
labels
```

```python
# decoding
le.inverse_transform([1,3,5,0])
```

3.표준화
- 평균 0. 표준편차 1인 표준정규분포로 전환

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
iris_std = ss.fit_transform(iris.data)
print('변환전:', iris.data[0])
print('변환후:',iris_std[0])
```

```python
#  모든 열이 포준정규분포로 변환되었음 확인
for i in range(4):
    print(iris_std[:, i].mean(), iris_std[:, i].std())
```

3. 로지스틱 회귀(Logistic Regression) 사례

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
#  변환하지 않은 데이터로 분류
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, stratify=iris.target, test_size=0.2, random_state=2023)

# max_iter 값이 100으로 설정되어 있어서 에러 발생
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
lr.get_params()
```

```python
lr = LogisticRegression(max_iter =500)
lr.fit(X_train, y_train)
```

```python
# 변환한 데이터로 분류
X_train, X_test, y_train, y_test=train_test_split(iris_std, iris.target, stratify=iris.target, test_size = 0.2, random_state=2023)
lr = LogisticRegression()
lr.fit(X_train, y_train)
```
