wine 분류

```python
from sklearn.datasets import load_wine
wine = load_wine()
```

데이터 전처리 및 탐색

```python
import numpy as np
import pandas as pd
df = pd.DataFrame(wine.data, columns = wine.feature_names)
df['target'] = wine.target
df.head()
```

```python
df.shape
```

```python
df.target.value_counts()
```

Train/ Test dataset 분리

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(wine.data, wine.target, stratify =wine.target, test_size =0.2, random_state=2023)
```

GridSearchCV 를 통한 최적의 분류기

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(random_state=2023)
params = {'max_depth': [2,5,8], 'min_samples_split': [2,3,4]}

grid_dt = GridSearchCV(dtc, params, scoring ='accuracy',cv=5)
grid_dt.fit(X_train, y_train)
```

```python
grid_dt.best_params_
```

```python
dtc = DecisionTreeClassifier(random_state=2023)
params = {'max_depth': [6,7,8,9,10], 'min_samples_split':[3,4,5,6]}
grid_dt =GridSearchCV(dtc, params, scoring='accuracy', cv = 5)
grid_dt.fit(X_train, y_train)
```

```python
grid_dt.best_params_
```

Best model로 예측/ 평가

```python
best_dt = grid_dt.best_estimator_
pred = best_dt.predict(X_test)
rf = pd.DataFrame({'y 실제값': y_test, '예측값': pred})
rf.head()
```

```python
best_dt.score(X_test, y_test)
```
